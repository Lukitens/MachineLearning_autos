```python
#Importamos las librerias
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use("bmh")
```

## EDA:

Leemos el dataframe

```python
df = pd.read_csv("ML_cars.csv")
df
```

### Usamos el describe all

```python
df.describe(include="all")
```

### Chequeamos nulos

```python
df.isna().mean()
```

### Chequeamos duplicados

```python
df.duplicated().mean()
```

#### Podemos ver que los datos están limpios y no tienen duplicados

### Vamos a hacer un grafico de dispersión para ver cuanto varian los precios de los autos en base a distintas características

#### Graficamos el precio en base a caballos de fuerza

```python
sns.scatterplot(data=df, x="price", y="horsepower")
plt.show()
```

#### Graficamos el precio en base al tamaño del motor

```python
sns.scatterplot(data=df, x="price", y="enginesize")
plt.show()
```

### Graficamos la cantidad de autos que son a gas y los que son de diesel en un grafico de barras

```python
#Agrupamos por el tipo de combustible
groupby_gas = df.groupby("fueltype")["aspiration"].count().reset_index(name="conteo")
#Graficamos
sns.barplot(data=groupby_gas, x="fueltype", y="conteo")
plt.show()
```

### Graficamos un conteo de la cantidad de autos en base a la cantidad de puertas

```python
#Agrupamos por la cantidad de puertas
groupby_puertas = df.groupby("doornumber")["aspiration"].count().reset_index(name="conteo")
#Graficamos
sns.barplot(data=groupby_puertas, x="doornumber", y="conteo")
plt.show()
```

### Hacemos el precio promedio en base a la estructura del auto

```python
#Hacemos el group by
groupby_estructura = df.groupby("carbody")["price"].mean().reset_index(name="precio_promedio")
#Graficamos
sns.barplot(data=groupby_estructura, x="carbody", y="precio_promedio")
plt.show()
```

### Vamos a graficar el precio en base a las revoluciones por minuto maximas

```python
#Graficamos un grafico de dispersión
sns.scatterplot(data=df, x="price", y="peakrpm")
plt.show()
```

### Vamos a extraer de la columna car names solo la marca del auto para poder ver que marca es la mas cara en promedio

```python
#Chequeamos los unicos de nombres de autos
df["CarName"].unique()
```

```python
#Extraemos los nombres de la marca
df["marca"] = df["CarName"].str.split().str[0]
```

```python
#Vemos los datos únicos y vemos que quedaron solo las marcas
df["marca"].unique()
```

### Vemos que hay varias marcas que son las mismas pero estan anotadas de otra forma, vamos a arreglarlo

```python
#Reemplazamos las marcas que esten mal escritas o abreviadas por el nombre entero y bien escrito
df["marca"].replace("vw", "volkswagen", inplace=True)
df["marca"].replace("vokswagen", "volkswagen", inplace=True)
df["marca"].replace("toyouta", "toyota", inplace=True)
df["marca"].replace("maxda", "mazda", inplace=True)
df["marca"].replace("porcshce", "porsche", inplace=True)
df["marca"].replace("Nissan", "nissan", inplace=True)
df["marca"].unique()
```

```python
#Agrupamos por marca
groupby_marca = df.groupby("marca")["price"].mean().reset_index(name="precio_promedio")
#Graficamos
sns.barplot(data=groupby_marca, x="precio_promedio", y="marca")
plt.show()
```
